# Numerical Data Poting 2
#### __relplot()__ (relational plot)
#### __scatterplot()__
#### __lineplot()__ (linear plot or line plot)


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn as rn

import warnings
warnings.filterwarnings('ignore')

```

```python
tips = sns.load_dataset('tips')
```

```python
tips.tail()
```

#### Check whether there is any relation between __total_bill__ and __tips data__

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(data= tips, x= 'total_bill', y = 'tip',hue = 'smoker', style='time')


## style = (time)
```

```python
sns.relplot(data= tips, x= 'total_bill', y = 'tip', style='size')

## Here style is equal to "size"
```

```python
tips['size'].value_counts()
```

```python
pd.options.display.max_rows = 10
```

```python
sns.relplot(data= tips, x= 'total_bill', y = 'tip', hue='size', palette='ch:r=-0.6, l=0.95')
```

```python
sns.relplot(data= tips, x= 'total_bill', y = 'tip', size='size')
```

# 3. Visualization Distributed of the Data

#### distplot()
#### kdeplot()
#### jointplot()
#### rugplot()

```python
import seaborn as sns
import pandas as pd
from numpy.random import randn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = randn(1000)
```

### __distplot()__

```python
sns.distplot(x, kde = True, hist= False)

## hist= False
```

```python
sns.distplot(x, kde = True, hist= True, color= 'red')

### hist= True
```

## __kdeplot()__

```python
sns.kdeplot(x, shade= True, color='green')
```
