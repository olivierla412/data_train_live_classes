## Let's do some Encoding...

```python
## converting gender column data into number

data['gender'] = data['gender'].replace({'Male':1,'Female':0})
```

```python
## converting Partner column data into number

data['Partner'] = data['Partner'].replace({'Yes':1,'No':0})
```

```python
## converting Dependents column data into number

data['Dependents'] = data['Dependents'].replace({'Yes':0,'No':1})
```

```python
## converting PhoneService column  data into number

data['PhoneService'] = data['PhoneService'].replace({'Yes':1,'No':0})
```

```python
## converting MultipleLines column data into number

data['MultipleLines'] = data['MultipleLines'].replace({'Yes':2,'No':1,'No phone service':0})
```

```python
## converting InternetService column data into number

data['InternetService'] = data['InternetService'].replace({'Fiber optic':2,'DSL':1,'No':0})
```

```python
## converting OnlineSecurity column data into number

data['OnlineSecurity'] = data['OnlineSecurity'].replace({'Yes':2,'No':1,'No internet service':0})
```

```python
## converting OnlineBackup column data into number

data['OnlineBackup'] = data['OnlineBackup'].replace({'Yes':2,'No':1,'No internet service':0})
```

```python
## converting DeviceProtection column data into number

data['DeviceProtection'] = data['DeviceProtection'].replace({'Yes':2,'No':1,'No internet service':0})
```

```python
## converting column TechSupport data into number

data['TechSupport'] = data['TechSupport'].replace({'Yes':2,'No':1,'No internet service':0})

```

```python
## converting column StreamingTV data into number

data['StreamingTV'] = data['StreamingTV'].replace({'Yes':2,'No':1,'No internet service':0})
```

```python
## converting column StreamingMovies data into number

data['StreamingMovies'] = data['StreamingMovies'].replace({'Yes':2,'No':1,'No internet service':0})
```

```python
## converting column Contract data into number

data['Contract'] = data['Contract'].replace({'Two year':2,'One year':1,'Month-to-month':0})
```

```python
## converting PaperlessBilling column  data into number

data['PaperlessBilling'] = data['PaperlessBilling'].replace({'Yes':1,'No':0})
```

```python
## converting column PaymentMethoddata into number

data['PaymentMethod'] = data['PaymentMethod'].replace({'Bank transfer (automatic)':3,'Credit card (automatic)':2,'Mailed check':1,'Electronic check':0}
                                                    )
```

```python
## converting Churn column data into number

data['Churn'] = data['Churn'].replace({'Yes':1,'No':0})

```

### __Skewness removal__

```python
## importing the libraries
from sklearn.preprocessing import PowerTransformer
```

```python
pt=PowerTransformer(method='yeo-johnson')
```

```python
## we will remove skewness form the features columns

features = new_data.drop('Churn',axis=1)
```

```python
X_power=pt.fit_transform(features)
```

```python
X_power
```

```python
## Now let's convert the result into a DataFrame

df=pd.DataFrame(X_power,columns=features.columns)
```

```python
df
```

```python
# let's check again the distribution in the feature columns.

plt.figure(figsize=(18,25))
plotnumber = 1

for column in df:
    if plotnumber <= 19:
        ax = plt.subplot(5,4,plotnumber)
        sns.distplot(df[column],bins=30,color='green')
        plt.xlabel(column,fontsize = 15,color='red')
    plotnumber+=1
plt.tight_layout()
```

### From here can see that our data look better than before, and it has less kewness.

```python
## let's check the classes in our target columns
new_data['Churn'].hist(grid=True,color='green')
plt.title('Churn(No ) VS Churn(Yes)')
plt.show()
```

### From this histogram we can see that the classes are imbalance and 'Churn(No ) is majority and 'Churn(Yes ) is the minority. So let's handling class imbalance using oversampling the minority class.

```python
## importang the libraries
from imblearn.over_sampling import SMOTE

SM = SMOTE()
```

```python
y = new_data['Churn']
```

```python
x_over , y_over = SM.fit_resample(df, y)
```

```python
## let's check the classes in our target columns
y_over.hist(grid=True,color='orange')
plt.title('Churn(No ) VS Churn(Yes)')
plt.show()
```

### Now we can see that our both classes of the label are balanced so we can build our model.

## __Before fit our data to the models ,let's see by visualize the relationship bw the features and the label usig stripplot() method.__

```python
plt.figure(figsize=(20,25), facecolor='white')
plot_no = 1

for column in x_over:
    if plot_no <= 19:
        ax = plt.subplot(5,4,plot_no)
        sns.stripplot(y_over, x_over[column])
    plot_no+=1
plt.tight_layout()
```

### From the above graphs, we can say that all the features are important and they have a  relationship with our label Churn (target column).

### __Now let's build the models__

```python
### Importing the libraries

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

### First with LogisticRegression and find the random state.

```python
max_accu = 0
max_RS = 0
for i in range(1,400):
    x_train, x_test, y_train,y_test = train_test_split(x_over,y_over,
                                                    test_size=30,
                                                    random_state=i)
    LR = LogisticRegression()
    LR.fit(x_train, y_train)
    predrf = LR.predict(x_test)
    acc = accuracy_score(y_test, predrf)
    if acc > max_accu:
        max_accu = acc
        max_RS = i
print('Best accuracy is :',max_accu, 'on Random_state',max_RS)
```

### __This code above help us to find the best random_state = 3 and accuracy = 96%__

```python
## let's build gain the model using the ramdom_state=3

x_train, x_test, y_train,y_test = train_test_split(x_over,y_over,
                                                    test_size=30,
                                                    random_state=3)
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
y_predlr = lr_model.predict(x_test)
print("The Accuracy is :",accuracy_score(y_test, y_predlr))
print('--'*20)
print(confusion_matrix(y_test,y_predlr))
print(classification_report(y_test,y_predlr))
```

### __We built the first model using LogisticRegression Algorithum and the accuracy is : 97%__

## Let's build a more models select the best one.

```python
## import the libraries

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC # Suport Vector Machine model
from sklearn.ensemble import GradientBoostingClassifier
```

```python
### DecisionTreeClassifier model.

dt = DecisionTreeClassifier(random_state=3)
dt.fit(x_train, y_train)
pred_dt = dt.predict(x_test)
print("The Accuracy is :",accuracy_score(y_test, pred_dt))
print('--'*20)
print(confusion_matrix(y_test, pred_dt))
print(classification_report(y_test,pred_dt))
```

### __DecisionTreeClassifier Accuracy is : 90%__

```python
## RandomForestClassifier

rf = RandomForestClassifier(3)
rf.fit(x_train, y_train)
pred_rf = rf.predict(x_test)

print("The Accuracy is :",accuracy_score(y_test, pred_rf))
print('--'*20)
print(confusion_matrix(y_test, pred_rf))
print(classification_report(y_test,pred_rf))
```

### __RandomForestClassifier Accuracy is : 83%__

```python
print('**'*30)
print('**'*30)
```

```python
# Suport Vector Machine model( SVC )

svc = SVC(random_state=3)

svc.fit(x_train, y_train)

pred_svc = svc.predict(x_test)

print("The Accuracy is :",accuracy_score(y_test, pred_svc))
print('--'*20)
print(confusion_matrix(y_test,pred_svc))
print(classification_report(y_test,pred_svc))
```

### __Suport Vector Machine model gives Accuracy = 97%__

```python
### GrandientBoostingClassifier model

gbdt_clf = GradientBoostingClassifier(random_state=3)
gbdt_clf.fit(x_train,y_train)
pred_gb = gbdt_clf.predict(x_test)
print("The Accuracy is :",accuracy_score(y_test, pred_gb))
print('--'*20)
print(confusion_matrix(y_test,pred_gb))
print(classification_report(y_test,pred_gb))
```
