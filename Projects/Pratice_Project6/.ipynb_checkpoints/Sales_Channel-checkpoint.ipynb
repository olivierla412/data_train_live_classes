```python
# importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
## loading the data
link ='https://raw.githubusercontent.com/dsrscientist/DSData/master/Advertising.csv'
data = pd.read_csv(link)
```

## Let's plot our dataset and see the distribution of each column.

```python
## plot a  distplot check the distribution

plt.figure(figsize=(18,15))
plotnumber = 1

for column in data:
    if plotnumber <= 4:
        ax = plt.subplot(2,2,plotnumber)
        sns.distplot(data[column],bins=30,color='purple')
        plt.xlabel(column,fontsize = 25,color='red')
    plotnumber+=1
plt.tight_layout()
```

## From distplot we see that our data is not distributed normally and we see some skewness form the newspaper feature.

```python
## let's plot the box check for outliers

plt.figure(figsize=(18,15))
plotnumber = 1

for column in data:
    if plotnumber <= 4:
        ax = plt.subplot(2,2,plotnumber)
        sns.boxplot(data[column],color='purple')
        plt.xlabel(column,fontsize = 25,color='red')
    plotnumber+=1
plt.tight_layout()
```

## From the plotbox we can see that onlt the newspaper column has outliers.

### let's remove the outliers using __zscore__

```python
from scipy import stats
import numpy as np
z= np.abs(stats.zscore(data))
print(z)
```

```python
print('-- Row nembers--   \n',np.where(z >3)[0])
print('-- columns numbers--   \n',np.where(z >3)[1])
```

```python
## save the rows in index_1 variable , we are interested in only rows 

index_1 =np.where(z > 3)[0]
```

```python
## Now let remove where record have  z > 3  (Row numbers )

data = data.drop(data.index[index_1])
```

```python
## Let's verify by plotting again the boxplot.

plt.figure(figsize=(18,15))
plotnumber = 1

for column in data:
    if plotnumber <= 4:
        ax = plt.subplot(2,2,plotnumber)
        sns.boxplot(data[column],color='purple')
        plt.xlabel(column,fontsize = 25,color='red')
    plotnumber+=1
plt.tight_layout()
```

#### we can see that the __outliers__ are removed form the newspaper column.

```python
## plot a  distplot check the distribution

plt.figure(figsize=(18,15))
plotnumber = 1

for column in data:
    if plotnumber <= 4:
        ax = plt.subplot(2,2,plotnumber)
        sns.distplot(data[column],bins=30,color='purple')
        plt.xlabel(column,fontsize = 25,color='red')
    plotnumber+=1
plt.tight_layout()
```

#### From the displot also we can see that the __sknewness__ from news papers is removed also.

## Now let's check the relationship between the label and features using scatter plot.

```python
## let's split our dataset into features and label.
x = data.drop('sales',axis=1)

y = data.sales
```

```python
plt.figure(figsize=(18,15), facecolor='white')
plotnumber = 1
for column in x :
    if plotnumber <=4 :
        ax = plt.subplot(2,2, plotnumber)
        plt.scatter(x[column],y, color ='purple')
        plt.xlabel(column, fontsize = 20)
        plt.ylabel('sales', fontsize = 20)        
    plotnumber+=1
plt.tight_layout()
```

### Plotting the Least Squares Line

```python


sns.pairplot(data, x_vars=['TV', 'radio','newspaper'], y_vars='sales', size=7, aspect=0.7, kind='reg')
```

#### From the scatter plot we can see that there is linear relationship between our label sales and the features. So we can say that our above __hypthesis is verified__.

#### We can see also the __TV__ and __Radio__ our stronger relationship with the label than the __newpaper__

## Now we going to build the models.

## LinearRegression()

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import metrics

```

```python
## let's standarized the features data 

scaler = StandardScaler()
X_scaler = scaler.fit_transform(x)
```

```python
#Split data into tain and test.
#Model will be build on training data and tested on test data.
x_train, x_test, y_train, y_test = train_test_split(X_scaler, y, test_size=0.25, random_state=20)
LR = LinearRegression()
LR.fit(x_train, y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
LR.score(x_test,y_test)
```

```python
print('\n*****Accuracy ****\n')

print(metrics.r2_score(y_test, y_pred))
```

```python
## let's build a second model
```

#### we build a first model using LinearRegression algorithum and the __accuracy is 90%__

### ************************************************************

## Let's build a second model Lasso(L1 form)

```python
## importing the libraries

from sklearn.linear_model import Ridge,RidgeCV,Lasso,LassoCV
```

```python
lass = Lasso()
lass.fit(x_train, y_train)
```

```python
lass.score(x_test, y_test)
```

```python
y_pred_lsso = lass.predict(x_test)
```

```python
print('\n*****Accuracy ****\n')

print(metrics.r2_score(y_test, y_pred_lsso))
```

#### we built third model using Lasso algorithum and the __accuracy is 80%.__

### ************************************************************

## __Let's build the Third model using Ridge regression algorithum.__

```python
ridge_model = Ridge()
ridge_model.fit(x_train, y_train)
```

```python
y_pred_rge = ridge_model.predict(x_test)
```

```python
print('\n*****Accuracy ****\n')

print(metrics.r2_score(y_test, y_pred_rge))
```

#### we built  third model using Ridge algorithum and the __accuracy is 90%.__

### Let's go for cross validation score and see if there no overfitting and wich model performe well.


```python
from sklearn.model_selection import cross_val_score
```

```python
# first model lr
scr = cross_val_score(LR,X_scaler,y,cv=7)
print('Cross Validation Score of LinearRegression model:',scr.mean())
```

```python
# first model lr
scr = cross_val_score(lass,X_scaler,y,cv=7)
print('Cross Validation Score of Lasso model:',scr.mean())
```

```python
# first model lr
scr = cross_val_score(ridge_model,X_scaler,y,cv=7)
print('Cross Validation Score of LinearRegression model:',scr.mean())
```

```python
print('\nLinearRegression model diff is:',90 -88)
print('\n Lasso Regression  model diff is:',80 -80)
print('\n Ridge Regression  diff is:',90 -80)
```

#### The minimum defference in accuracy and cross validation score is for lasso Regression  model, So it is our best model.
