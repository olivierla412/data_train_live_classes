### Outlier Dectection Formula

##### __Higher side ===> Q3 + (1.5 * IQR)__
##### __lower side ===> Q1 - (1.5 * IQR)__
##### __IQR = Inter Quantile Rate = Q3 - Q1__

```python
## Find the IQR ( Inter Quantile Range) to identify outliers

# 1st quantile
q1 = df.quantile(0.25)

# 3rd quantile
q3 = df.quantile(0.75)

#IQR 
iqr = q3 - q1
```

```python
#3 - Outliers in { height, weight ,ap_hi, ap_lo }
```

```python
# Validation one outlier, max size I should take

height_hi = (q3.height + (1.5 * iqr.height))
height_hi
# this is the max_vales ,so all all record greater than this values in the column height should be dropped.
```

### Let's remove the Outlier Higher side

```python
import numpy as np
```

```python
remove_index = np.where(df['height'] > height_hi)
remove_index
```

```python
## just to verify at index 82 values 

df['height'].head(39)
# at index(328) values = 195 which is greater than 186.5 my max size , so verified
```

```python
# This is the first

df = df.drop(df.index[remove_index])
df.reset_index(drop=True)
```

```python
remove_index = np.where(df['ap_lo'] > (q3.ap_lo + (1.5 * iqr.ap_lo)))
df= df.drop(df.index[remove_index])
df.reset_index(drop=True)
```

## This must applied to only __Continous Labels__ !!!!! __zscore__

```python
# with std 3 let's see the stats
# from scipy import stats
from scipy.stats import zscore
```

```python
df.head(8)
```

### Only thses column or label have continous data : __height,weight,ap_hi,ap_lo__

```python
z_score = zscore(df[['age','height','ap_hi','ap_lo']])
abs_z_score = np.abs(z_score)
filtering_entry = (abs_z_score < 3).all(axis =1)
df = df[filtering_entry]
```

```python
# Let's build the model now

X = df.drop('cardio',axis=1)
Y = df.cardio
```

##### Let's check if there is any multicollinearity

```python
scale = StandardScaler()
X_scaler = scale.fit_transform(X)
```

```python
## variance_inflation_factor (vif)

vif  = pd.DataFrame()
vif['score'] = [variance_inflation_factor(X_scaler,i) for i in range(X_scaler.shape[1])]
vif['feature'] = X.columns
print(vif)
```

#### There is not score greater than 5 __( score < 5 )__, so we say that there no multicollinearity

### RandomForestClassier

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaler,Y,train_size=0.25,random_state=355)

clf = RandomForestClassifier()
params = {
    'n_estimators':[10,12,14,17],
    'criterion':['entropy'],
    'max_depth':range(5,15,1),
    'min_samples_split':[10],
    'min_samples_leaf':[5,7,8]
}

```

```python
grid= GridSearchCV(clf,param_grid=params)

grid.fit(x_train,y_train)
print('The best parameters are: ',grid.best_params_)
```

#### Reinstantiating with the best params


```python
clf = grid.best_estimator_
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
rf_conf_mat = confusion_matrix(y_test,y_pred)

print('confusion_matrix ==>','\n',rf_conf_mat)

print('Accuracy of rf ===>',accuracy_score(y_test,y_pred))
```

```python
## let's check the report
report = classification_report(y_test,y_pred)
print(report)
```
